```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import seaborn as sns
data = pd.read_csv(r"C:\Users\lesar\projects\coderhouse\hrdata.csv", sep=',' ,encoding='latin-1')
#Se Realizan las estadisticas 
data2 = data.describe()
print(data2)
```

```python
#Se realiza ek histograma con 30 bins para obtener varios rangos
salario = data['Salary']
salario2 = salario.value_counts()
sns.set(rc={'figure.figsize':(16,10)})
sns.histplot(data=salario, bins=30)
```

```python
#Se realiza un grfico de violn con los salarios, discriminados por gnero
sns.violinplot(data['Sex'], data['Salary'])
```

```python
#Se grafica la serie de tiempo correspondiente a la fecha de contratacin de manera cronolgicamente organizada
data4 = pd.DataFrame(data['DateofHire'].value_counts())
sns.lineplot(data=data4.sort_index())
plt.xticks(rotation=90)
plt.title('seaborn-matplotlib example')
plt.show()
```
